# 0. Import thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Đọc dữ liệu House_Prices

```python
# Nạp dữ liệu từ file CSV
houseprices_data = pd.read_csv("/Users/mac/Desktop/PTDL_GK/LAB1_PTDL/phan_tich_don_bien_hai_bien/bai_toan_1_don_bien/nhiem_vu_2_house_prices/data/HousingPrices-Amsterdam-August-2021.csv")

# Chỉ lấy các cột cần phân tích
houseprices_data = houseprices_data[['Zip', 'Price', 'Area', 'Room']]
```

# 2. Phân tích đơn biến bằng Boxplot
# Case 1: Basic

```python
sns.boxplot(data=houseprices_data, x=houseprices_data["Price"])
plt.show()
```

# Case 2: Advanced

```python
plt.figure(figsize=(12,6))
ax = sns.boxplot(data=houseprices_data, x=houseprices_data["Price"])
ax.set_xlabel('House Prices in millions', fontsize=15)
ax.set_title('Univariate Analysis of House Prices', fontsize=20)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

# 3. Phân tích đơn biến bằng Violin Plot
# Case 1: Basic

```python
sns.violinplot(data=houseprices_data, x=houseprices_data["Price"])
plt.show()
```

# Case 2: Advanced

```python
plt.figure(figsize=(12,6))
ax = sns.violinplot(data=houseprices_data, x=houseprices_data["Price"])
ax.set_xlabel('House Prices in millions', fontsize=15)
ax.set_title('Univariate Analysis of House Prices', fontsize=20)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```
