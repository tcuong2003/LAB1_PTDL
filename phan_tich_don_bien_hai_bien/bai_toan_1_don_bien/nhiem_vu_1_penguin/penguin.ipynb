# 0. Import thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Đọc dữ liệu Penguins

```python
penguins_data = pd.read_csv("/Users/mac/Desktop/PTDL_GK/LAB1_PTDL/phan_tich_don_bien_hai_bien/bai_toan_1/nhiem_vu_1_penguin/data/penguins_lter.csv")
penguins_data.columns = penguins_data.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
penguins_data = penguins_data[['species','culmen_length_mm']]
```

# 2. Phân tích đơn biến với Histogram (culmen_length_mm)
# Case 1: Basic histogram

```python
sns.histplot(data=penguins_data, x="culmen_length_mm")
plt.show()
```

# Case 2: Advanced histogram

```python
plt.figure(figsize=(12,6))
ax = sns.histplot(data=penguins_data, x="culmen_length_mm", bins=20, kde=True)
ax.set_xlabel('Culmen Length in mm', fontsize=15)
ax.set_ylabel('Count of records', fontsize=15)
ax.set_title('Univariate analysis of Culmen Length', fontsize=20)
plt.show()
```

# 3. Phân tích đơn biến với Bar chart (species)
# Case 1: Basic bar chart

```python
sns.countplot(data=penguins_data, x="species")
plt.show()
```

# Case 2: Advanced bar chart

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(data=penguins_data, x="species", hue="species", palette="Set2", legend=False)
ax.set_xlabel('Penguin Species', fontsize=15)
ax.set_ylabel('Count of records', fontsize=15)
ax.set_title('Univariate analysis of Penguin Species', fontsize=20)
plt.show()
```

# 4. Phân tích đơn biến với Pie chart (species)

```python
penguins_group = penguins_data.groupby('species').count().reset_index()
```

# Case 1: Basic pie chart

```python
plt.pie(penguins_group["culmen_length_mm"], labels=penguins_group['species'])
plt.show()
```

# Case 2: Advanced pie chart

```python
cols = ['g', 'b', 'r']
plt.pie(
    penguins_group["culmen_length_mm"],
    labels=penguins_group['species'],
    colors=cols,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Univariate Analysis of Species', fontsize=15)
plt.show()
```
