# 0. Import thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Đọc dữ liệu Penguins_lter

```python
# Đọc dữ liệu từ file CSV
penguins_data = pd.read_csv("/Users/mac/Desktop/PTDL_GK/LAB1_PTDL/phan_tich_don_bien_hai_bien/bai_toan_2_da_bien/nhiem_vu_1_penguin2z/data/penguins_lter.csv")

# Chọn các cột cần thiết để phân tích hai biến
penguins_data = penguins_data[['Species', 'Culmen Length (mm)', 'Body Mass (g)', 'Sex']]
```

# 2. Phân tích hai biến bằng Scatter Plot
# Case 1: Basic

```python
sns.scatterplot(data=penguins_data, 
                x=penguins_data["Culmen Length (mm)"], 
                y=penguins_data["Body Mass (g)"])
plt.show()
```

# Case 2: Advanced

```python
plt.figure(figsize=(12,6))
ax = sns.scatterplot(data=penguins_data,
                     x=penguins_data["Culmen Length (mm)"],
                     y=penguins_data["Body Mass (g)"],
                     hue=penguins_data["Species"])  # tô màu theo loài
ax.set_title('Bivariate Analysis of Culmen Length and Body Mass', fontsize=20)
ax.set_xlabel('Culmen Length (mm)', fontsize=14)
ax.set_ylabel('Body Mass (g)', fontsize=14)
plt.show()
```

# 3. Phân tích hai biến bằng bảng Crosstab (Two-Way Table)

```python
pd.crosstab(index=penguins_data['Species'], columns=penguins_data['Sex'])
```

# 4. Phân tích hai biến bằng Pivot Table

```python
import numpy as np
pd.pivot_table(
    penguins_data,
    values='Culmen Length (mm)',
    index='Species',
    aggfunc='mean'
)
```

# 5. Phân tích đa biến bằng Pairplot

```python
sns.pairplot(data=penguins_data,
             vars=['Culmen Length (mm)', 'Body Mass (g)'],
             hue='Species')
plt.show()
```
